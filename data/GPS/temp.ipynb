```python
margin = 0.002

checked_opacity = 0.15

max_lat = max(lats) + margin
min_lat = min(lats) - margin
max_lon = max(lons) + margin 
min_lon = min(lons) - margin

for i in range(len(lats)): 
  if points_to_check[names[i]]: opacity = 1 
  else: opacity = checked_opacity
  plt.scatter(lons[i], lats[i], label=names[i], alpha = opacity)
  plt.annotate(names[i], (lons[i], lats[i]), textcoords="offset points", xytext=(0, 10), ha='center', size=10, alpha = opacity)

# Watch cave only
plt.scatter(13.767311, 46.25069, label="WATCH CAVE", alpha = opacity)
plt.annotate("WATCH CAVE", (13.767311, 46.25069), textcoords="offset points", xytext=(0, 10), ha='center', size=10, alpha = opacity)

# Lines for locating caves
# name_to_locate = "VILA1"
# plt.hlines([data[names.index(name_to_locate)][1]], [min(lons) - margin], [max(lons) + margin])
# plt.vlines([data[names.index(name_to_locate)][2]], [min(lats) - margin], [max(lats) + margin])

grid_num = 50

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.xticks(np.linspace(min(lons), max(lons), grid_num))
plt.yticks(np.linspace(min(lats), max(lats), grid_num))
plt.grid()
plt.xlim(min_lon, max_lon)
plt.ylim(min_lat, max_lat)

plt.interactive(True)
```
